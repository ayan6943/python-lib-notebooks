**Matplotlib is a Python library used for creating static, animated, and interactive plots. It is highly customizable and widely used for data visualization.**

```python
import matplotlib.pyplot as plt
import numpy as np

# Line Plot
X = [1, 2, 3]
Y = [1, 2, 3]
plt.plot(X, Y,marker="*",lw=2,ls="-",c='red',label="line",ms=6,markeredgecolor='black',mfc="blue")
plt.plot([1,3,6],X,marker="*",lw=2,ls="-",c='blue',label="line",ms=6,markeredgecolor='black',mfc="red")
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()

# Scatter Plot
X = np.random.rand(10)
Y = np.random.rand(10)  # Add Y-values
plt.scatter(X,Y,marker="1",color="red",label='one')
plt.scatter([1,2,3,4,5,6,7,8,9,0],Y,marker="*",color="blue",label='two')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

**Visualising Error**

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

# Fixed error for each point (same for simplicity)
error = [0.5, 0.3, 0.4, 0.2, 0.6]

# Plot with error bars
plt.errorbar(x, y, yerr=error, fmt='o',ecolor='red', capsize=5)

# Add labels
plt.title("Simple Error Bar Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid(True)
plt.show()
```

**Density and Contour Plots**

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 2D grid of values
data = np.random.rand(100, 100)

plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.title("Density Plot (Heatmap)")
plt.colorbar(label="Value")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create X and Y grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# A simple function of X and Y
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot contour lines
plt.contour(X, Y, Z, levels=20, cmap='plasma')
plt.title("Contour Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar(label="Value")
plt.show()
```

```python
from scipy.stats import multivariate_normal

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = multivariate_normal([0, 0], [[1, 0.5], [0.5, 1]]).pdf(np.dstack((X, Y)))

plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar()
plt.title("Contour Plot")
plt.show()
```

**Subplots**

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].plot(x, y1, 'r--')
axs[0].set_title("Sine Wave")

axs[1].plot(x, y2, 'b-.')
axs[1].set_title("Cosine Wave")

plt.tight_layout()
plt.show()
```

**3-D Plotting**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure object
fig = plt.figure()

# Add a 3D subplot to the figure
ax = fig.add_subplot(111, projection='3d')

# Generate data points for the X and Y axes using linspace
x = np.linspace(-5, 5, 50)  # 50 points between -5 and 5
y = np.linspace(-5, 5, 50)  # 50 points between -5 and 5

# Create a meshgrid from the X and Y data to generate a grid of X, Y values
X, Y = np.meshgrid(x, y)

# Define the Z values for each point in the grid (function Z = X^2 + Y^2)
Z = X**2 + Y**2

# Plot the 3D surface plot with the X, Y, and Z values
ax.plot_surface(X, Y, Z, cmap='viridis')  # cmap specifies the color map
# Set title for the plot
plt.title("Simple 3D Surface Plot")

# Show the plot
plt.show()
```

**Geographic data with Basemap**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Create a figure
fig = plt.figure(figsize=(10, 7))

# Create a Basemap instance
m = Basemap(projection='ortho', lat_0=0, lon_0=0)

# Draw coastlines and countries
m.drawcoastlines()
m.drawcountries()

# Define the locations (latitude and longitude) of cities
cities = {
    'New York': (-74.006, 40.7128),    # Longitude, Latitude
    'London': (-0.1278, 51.5074),
    'Tokyo': (139.6917, 35.6762),
}

# Plot the cities on the map
for city, (lon, lat) in cities.items():
    x, y = m(lon, lat)  # Convert longitude and latitude to map projection coordinates
    m.plot(x, y, marker='D', color='red', markersize=8)  # Plot the city as a red diamond
    plt.text(x, y, city, color='black', fontsize=12, ha='center')  # Add city name

# Show the plot
plt.title("World Map with Cities")
plt.show()
```

```python
from matplotlib import pyplot as plt

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
plt.title("Contour Plot")
plt.show()
```

# Visualising with Seaborn

```python
import seaborn as sns
import pandas as pd

# Load built-in dataset
df = sns.load_dataset('iris')

# Pairplot
sns.pairplot(df, hue='species')
plt.suptitle("Seaborn Pairplot", y=1.02)
plt.show()

# Boxplot
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title("Boxplot: Sepal Length by Species")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Sample dataset
x = np.linspace(0, 10, 100)
y = np.sin(x)
data = np.random.normal(loc=50, scale=15, size=200)

# 1. Line chart using plot()
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Sine wave', color='blue')
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()

# 2. Histogram to show distribution
plt.figure(figsize=(6, 4))
plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.title('Histogram - Distribution of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 3. Scatter plot to find patterns
np.random.seed(0)
x_scatter = np.random.rand(50)
y_scatter = 2 * x_scatter + np.random.normal(0, 0.1, 50)

plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, color='green')
plt.title('Scatter Plot - Pattern in Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

# 4. Box plot for statistical summary
data_box = pd.DataFrame({
    'Group A': np.random.normal(60, 10, 100),
    'Group B': np.random.normal(55, 15, 100),
    'Group C': np.random.normal(70, 8, 100)
})

plt.figure(figsize=(6, 4))
sns.boxplot(data=data_box)
plt.title('Box Plot - Statistical Summary')
plt.grid(True)
plt.show()

# 5. Legend to identify color patterns
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
y2 = np.cos(x1)

plt.figure(figsize=(6, 4))
plt.plot(x1, y1, label='Sine', color='red')
plt.plot(x1, y2, label='Cosine', color='blue')
plt.title('Legend Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title="Functions")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulating product prices with a normal distribution
np.random.seed(42)

# Generate 200 random product prices for each category
electronics_prices = np.random.normal(loc=500, scale=100, size=200)  # Mean = 500, SD = 100
clothing_prices = np.random.normal(loc=50, scale=20, size=200)      # Mean = 50, SD = 20
groceries_prices = np.random.normal(loc=20, scale=10, size=200)     # Mean = 20, SD = 10

# Create a box plot for each category
plt.figure(figsize=(8, 6))
plt.boxplot([electronics_prices, clothing_prices, groceries_prices],
            labels=['Electronics', 'Clothing', 'Groceries'])

# Add titles and labels
plt.title('Product Price Distribution')
plt.xlabel('Category')
plt.ylabel('Price')
plt.show()
```
